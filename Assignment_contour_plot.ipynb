```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
```

```python
x1 = np.linspace(-3.0, 3.0, 100)
x2 = np.linspace(-3.0, 3.0, 100)
```

```python
xx1, xx2 = np.meshgrid(x1,x2)
```

```python

#Linear Function
fig = plt.figure(figsize=(10,10))

ax = fig.gca(projection='3d')

ax.plot_surface(xx1,xx2, xx1+xx2 +10,cmap=cm.coolwarm)

plt.show()
plt.contourf(xx1,xx2, xx1+xx2 +10)

plt.colorbar()
plt.show()
plt.contour(xx1,xx2, xx1+xx2 +10)
plt.colorbar()
plt.show()
```
